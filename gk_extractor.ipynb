```python
c_d={}

with open('TlkPersonaChatRus/tolokapersonachat.jsonl', 'r') as inp:
    for line in tqdm.tqdm(inp):
        line = json.loads(line)
        for k in line:
            if k=='responce':
                for word in line[k].split():
                    word = [l for l in word if l.isalpha()]
                    word = ''.join(word)
                    word = morph.parse(word)[0].normal_form
                    try:
                        c_d[word]+=1
                    except KeyError:
                        c_d[word]=1
            elif k=='label':
                pass
            else:
                for sent in line[k]:
                    for word in sent.split():
                        word = [l for l in word if l.isalpha()]
                        word = ''.join(word)
                        word = morph.parse(word)[0].normal_form
                        word = morph.parse(word)[0].normal_form
                        try:
                            c_d[word]+=1
                        except KeyError:
                            c_d[word]=1
        with open('TlkPersonaChatRus/count_dict.json', 'w') as out:
            out.write(json.dumps(c_d, ensure_ascii=False))
            
print(c_d)
```

```python
L={}
alp=list('qwertyuiopasdfghjklzxcvbnm')
with open('TlkPersonaChatRus/tolokapersonachat_gk.jsonl', 'w') as out:
    with open('TlkPersonaChatRus/tolokapersonachat.jsonl', 'r') as inp:
        for line in inp:
            line = json.loads(line)

            responce_ = line['responce']
            responce = responce_.split()
            ngram=[]
            for word in responce:
                word = [l for l in word if l.isalpha() and l not in alp]
                word = ''.join(word)
                word = morph.parse(word)[0]
                ngram.append(word.normal_form)
                
            responce = ngram

            persona_ = line['persona']
            persona=[]
            for p in persona_:
                p = p.split()
                ngram=[]
                for word in p:
                    word = [l for l in word if l.isalpha() and l not in alp]
                    word = ''.join(word)
                    word = morph.parse(word)[0]
                    if word.tag.POS=='NOUN':
                        ngram.append(word.normal_form)
                    elif word.tag.POS=='INFN':
                        ngram.append(word.normal_form)
                persona.append(ngram)

            golden_p=[]
            for i, p in enumerate(persona):
                for wp in p:
                    for wr in responce:
                        if wp in wr or wr in wp:
                            if i not in golden_p:
                                golden_p.append(i)

            try:
                L[len(golden_p)]+=1
            except:
                L[len(golden_p)]=1

            line['response'] = line['responce']
            del line['responce']
            del line['label']
            line['golden'] = golden_p
            line = json.dumps(line, ensure_ascii=False)
            out.write(line+'\n')
print(L)
```
